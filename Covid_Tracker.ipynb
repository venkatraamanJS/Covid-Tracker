```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.head()
```

```python
df=df.iloc[:-1,1:]
south_India_df = df[(df['state_name']=='Andhra Pradesh') | (df['state_name']=='Tamil Nadu') | (df['state_name']=='Goa') | (df['state_name']=='Kerala') | (df['state_name']=='Telangana')]
south_India_df["state_name"].replace({"Andhra Pradesh": "AP", "Tamil Nadu": "TN"}, inplace=True)
South_states = np.arange(len(south_India_df['state_name'])) 
```

### Total Positive Cases Based On Tamil nadu, kerala, karnataka, Andhra pradesh 

```python
figure(num=None, figsize=(12, 12), dpi=1024, facecolor='w', edgecolor='k') 

plt.subplot(221)
plt.title('Confirmed')
plt.barh(South_states,south_India_df['positive'], align='center', alpha=0.5,color=(1, 0.02, 0.22))   
plt.xlabel('Confirmed Number of Cases')
plt.yticks(South_states, south_India_df['state_name'])  
plt.xlim(1,max(south_India_df['positive'])) 

plt.subplot(222)
plt.title('Active')
plt.barh(South_states,south_India_df['active'], align='center', alpha=0.5,color=(0, 0.4, 1))
plt.xlabel('Active Number of Cases')
plt.yticks(South_states, south_India_df['state_name'])  
plt.xlim(1,max(south_India_df['active'])) 

plt.subplot(223)
plt.title('Recovered')
plt.barh(South_states,south_India_df['cured'], align='center', alpha=0.5,color=(0.1, 0.6, 0.2))
plt.xlabel('Recovered Number of Cases')
plt.yticks(South_states, south_India_df['state_name'])  
plt.xlim(1,max(south_India_df['cured']))

plt.subplot(224)
plt.title('Deceased')
plt.barh(South_states,south_India_df['death'], align='center', alpha=0.5,color=(0.4, 0.4, 0.49))
plt.xlabel('Death Number of Cases')
plt.yticks(South_states, south_India_df['state_name'])  
plt.xlim(1,max(south_India_df['death']))

plt.show()  
plt.savefig('case_study.png')
```

```python
figure(num=None, figsize=(12, 12), dpi=1024, facecolor='w', edgecolor='k')

#  subplot #1
plt.subplot(221)
plt.title('Confirmed')
colors = sns.color_palette('pastel')[0:5]
plt.pie(south_India_df['positive'], labels = south_India_df['state_name'], colors = colors, autopct='%.0f%%')

#  subplot #2
plt.subplot(222)
plt.title('Active')
colors = sns.color_palette('pastel')[0:5]
plt.pie(south_India_df['active'], labels = south_India_df['state_name'], colors = colors, autopct='%.0f%%')


#  subplot #3
plt.subplot(223)
plt.title('Recovered')
colors = sns.color_palette('pastel')[0:5]
plt.pie(south_India_df['cured'], labels = south_India_df['state_name'], colors = colors, autopct='%.0f%%')

#  subplot #4
plt.subplot(224)
plt.title('Deceased')
colors = sns.color_palette('pastel')[0:5]
plt.pie(south_India_df['death'], labels = south_India_df['state_name'], colors = colors, autopct='%.0f%%')

plt.show()
plt.savefig('case_study.png')
```
